### Задание 3. Работа с пропусками

Рассчитать количество пропусков для всех выбранных столбцов.

Принять и обосновать решение о методе заполнения пропусков по

каждому столбцу на основе рассчитанных статистик и возможной

```python
df_horse.isna().mean() * 100
# количество пропусков в процентах
```

```python
lived_young = df_young[df_young['outcome'] == 1]
lived_young

# предполагаю, что параметры организмов выживших жеребят 
# отличаются от погибших.
# В свою очередь, разделяю жеребят на живых и покойных (умерших и усыпленных)
```

```python
lived_young.describe()
```

```python
lived_young_m = lived_young.copy()
lived_young_m[['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain']] \
    = lived_young_m[['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain']] \
    .fillna(lived_young[['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain']].median())

lived_young_m

# заменю пропуски на медианные значения
```

```python
died_young = df_young[df_young['outcome'] != 1]
died_young
```

```python
died_young_m = died_young.copy()           
died_young_m[['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain']] \
    = died_young_m[['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain']] \
    .fillna(died_young[['rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'pain']].median())

died_young_m
```

```python
f_young = pd.concat([lived_young_m, died_young_m])     
f_young

# объединю в один датафрейм заполненные таблицы 
```
